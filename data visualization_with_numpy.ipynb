```python
import numpy as np
import matplotlib.pyplot as plt
```

# Line plot:-

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Line plot")
plt.plot(x,y)
plt.show()
```

# custom Line Graph

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Line plot")
plt.plot(x,y,'o', linestyle='--',color='green',alpha=.5)
plt.show()
```

# Mult plot graph

```python
x = np.linspace(0,10,50)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,'o', linestyle='--',color='green', markersize = 5,alpha=.5,label='x-axis')
plt.plot(x,z,'o', linestyle='--',color='red',  markersize = 5,alpha=.5,label='y-axis')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Multi plot")
plt.legend()
plt.show()
```

# Scatter plot

```python
x = np.random.rand(500)
y = np.random.rand(500)
plt.scatter(x,y,marker='o',color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter plot')
plt.show()
```

# Custom Scatter graph:-

```python
x = np.random.rand(500)
y = np.random.rand(500)
z = np.random.rand(500)
plt.scatter(x,y,marker='o',cmap='viridis',c=z)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter plot')
plt.colorbar()
plt.show()
```

# Scatter graph with multiple data:-

```python
x = np.random.rand(200)+1
y = np.random.rand(200)+1

x1 = np.random.rand(200)-1
y1 = np.random.rand(200)-1

plt.scatter(x,y,marker='o',)
plt.scatter(x1,y1,marker='o',)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter plot')
plt.colorbar()
plt.show()
```

# Hexbinning Scatter Plot:

```python
x = np.random.rand(500000)
y = np.random.rand(500000)
plt.hexbin(x,y,gridsize=20)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Hexbin plot')
plt.colorbar()
plt.show()
```

# Basic Bar Graph:

```python
x = np.array(['A','B','C','D'])
y = np.array([10,20,30,40])
plt.bar(x,y,color='yellow')
plt.title('Bar graph')
plt.show()
```

# Stacked Bar Graph: 

```python
x = np.random.rand(10,5)

plt.bar(np.arange(10), x[:,0] +x[:,1]+x[:,2]+x[:,3]+x[:,4],label='D-5')
plt.bar(np.arange(10), x[:,0] +x[:,1]+x[:,2]+x[:,3],label='D-4')
plt.bar(np.arange(10), x[:,0] +x[:,1]+x[:,2],label='D-3')
plt.bar(np.arange(10), x[:,0] +x[:,1],label='D-2')
plt.bar(np.arange(10), x[:,0] ,label='D-1')
plt.legend()
plt.title("Stacked bar graph")
plt.show()
```

# Grouped Bar Graph:

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.array([[10,20,30],
                 [40,50,60],
                 [70,80,90]])

width = .25
x = np.arange(3)

plt.bar(x-width, data[:,0],.25,label = 'D-1')
plt.bar(x        , data[:,1],.25,label = 'D-2')
plt.bar(x+width , data[:,2],.25, label = 'D-3')

plt.legend()
plt.show()
```

# Basic Histogram:

```python
x = np.random.randn(100000000)
plt.hist(x, bins=1000)
plt.show()
```

```python
x = np.random.randn(100000000)
plt.hist(x, bins=100,cumulative=True)
plt.show()
```

```python
x = np.random.randn(100000000)
y = np.random.randn(100000000)+2
plt.hist(x, bins=1000,label='D-1',color='yellow')
plt.hist(y, bins=1000,label='D-2',color='blue')
plt.title('Multi Histogram')
plt.legend()
plt.show()
```

#  Box Plot:

```python
x = np.random.rand(10,5)

plt.boxplot(x,labels=['D-5','D-4','D-3','D-2','D-1'],vert=True)
plt.title("Box plot")
plt.show()
#vert is for horz or vertical of boxplots..if True = horz else Flase=vertical
```

# Pie Chart:

```python
import numpy as np
import matplotlib.pyplot as plt
labels = ['Python', 'C++', 'Java', 'JavaScript', 'PHP']
sizes =  [21.3,        14.6 , 46.1,     8.5,       7.7]
explode = (0.13,        0,    0,       0,           0)
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']


plt.title("Programming Language Popularity")
plt.pie(sizes, colors=colors, labels=labels, autopct='%1.2f%%', shadow=True,explode=explode,startangle=360)
#explode for break the pie chart as show python is not attached.

plt.axis('equal')
plt.legend(title="Languages", loc="upper left")

plt.show()
```
